# CSV Files Type

```python
from io import StringIO, BytesIO
```

```python
data= ("col1,col2,col3\n"
      "x,y,1\n"
      "a,b,2\n"
      "k,z,3\n")
```

```python
df3=pd.read_csv(StringIO(data))
```

```python
pd.read_csv(StringIO(data), usecols=["col1","col3"])
```

```python
df3.head()
```

```python
df3.to_csv("Test2.csv")
```

## Specifying the Column data types

```python
data= ("a,b,c,d\n"
        "1,2,3,4\n"
      "5,6,7,8\n"
      "9,10,11,12")
```

```python
print(data)
```

```python
df5=pd.read_csv(StringIO(data), dtype=float)
```

```python
df5.head()
```

```python
df5["a"][0]
```

```python
type(df5["a"][0])
```

```python
df6=pd.read_csv(StringIO(data), dtype={"a":int, "b":float, "c": int})
```

```python
df6
```

```python
type(df6["b"][1])
```

```python
# Check the data types
df6.dtypes
```

```python
data= ("index,a,b,c\n"
      "4,apple,bat,5.7\n"
      "8,orange,cow,10.0\n")
```

```python
df_csv=pd.read_csv(StringIO(data), index_col=1)
```

```python
df_csv
```

```python
type(df_csv)
```

```python
pd.read_csv(StringIO(data), index_col=False)
```

```python
data= ("a,b,c\n"
      "4,apple,bat\n"
      "8,orange,cow\n")
```

```python
pd.read_csv(StringIO(data), index_col=False)
```

```python
pd.read_csv(StringIO(data), usecols=["b", "c"])
```

```python
pd.read_csv(StringIO(data), usecols=["b", "c"], index_col=False)
```

```python
pd.read_csv(StringIO(data), index_col=False)
```

### Quoting and escape Characters. Very useful in NLP.

```python
data= ( 'a,b\n "hello,\\"Bob\\", nice to see you",5')
pd.read_csv(StringIO(data), escapechar="\\")
```
