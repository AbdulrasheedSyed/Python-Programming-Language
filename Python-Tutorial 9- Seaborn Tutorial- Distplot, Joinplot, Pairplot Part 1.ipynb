## Corelation with Heatmap

A corelation heatmap uses coloured cells, typically in a monocromatic scale, to show a 2d co-relation matrix(table)
between two discrete dimentions or even types. It is very important in feature selection.


```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Joint Plots

 A Joint plot allows to study two different numeric variables.The contral chart display the correlation.
 it is usually a scatterplot, a hexibin plot, a 2d Histogram, or a 2D density plot.

## Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='hex',color='b')
```

```python
sns.jointplot(x='tip', y= 'total_bill', data= df, kind= 'reg', color='b')
```

## Pair Plot

```python
 A pair plot is also know as scatter plot. in which one variable in the same data row is match with another variable's value.
 like this : pairplots are just elaborations on this, showing all variables paired with all other variables. 
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue= 'sex')
```

```python
sns.pairplot(df, hue= 'smoker')
```

```python
df["smoker"].value_counts()
```

## Dist Plots

 Dist plots help us to check the distribution of columns features.

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=50)
```

## Violinplot

- violin plot help us to see both the distribution of data in terms of kurnel dencity estimation and boxplot.

```python
sns.violinplot('total_bill', 'day', data=df, palette='rainbow')
```
