# MatplotLib Tutorial

Matplotlib is a plotting Library for the python Programming language and its numerical mathematics extension is NumPy.
It provides an object oriented API for the embedding plots into applications using general purpose GUI toolkits like Tkinter,
wxPython,Qt, or GTK+


Some of the major Pros of MatplotLib are: 
- Generally easy to get start with simple plots.
- Support for custom lables and texts
- Great controle of every element in a figure
- High- Quality output in many forms
- very custamizable in general



```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

## Scatter plot

```python
# Plotting Using matplotlib
# plt scatter 
plt.scatter(x,y,c= "g" )
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(" Graphic 2D")
plt.savefig("Test1.png")
```

## Plt Plot

```python
# Plt Plot
y=x*x
```

```python
plt.plot(x,y,'r*', linestyle="dashed", linewidth=2, markersize=20)
plt.xlabel ("X-Axis")
plt.ylabel("Y-axix")
plt.title(" Line 2D")
```

## Creating Sub plots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'ro--')
plt.subplot(2,2,2)
plt.plot(x,y,'go--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo--')
plt.subplot(2,2,4)
plt.plot(x,y,'yo--')
```

```python
x=np.arange(0,10)
y=x*3+10
plt.plot(x,y,'bo--')
plt.xlabel ("X-Axis")
plt.ylabel("Y-axix")
plt.title(" Line 2D")
```

## Sign Curves

```python
np.pi
```

```python
x= np.arange(0, 5*np.pi, 0.1)
y=np.sin(x)
plt.plot(x,y, 'go--')
plt.title("Sin Curve")
```

```python
x= np.arange(0, 5*np.pi, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin, 'r*--')
plt.subplot(2,1,2)
plt.plot(x,y_cos, 'g*--')
```

## Bar Plots

```python
x=[2,8,10]
y=[11,16,9]
```

```python
x2= [3,9,11]
y2=[6,15,7]
```

```python
plt.bar(x,y,color='b')
plt.bar(x2,y2,color='g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Plots")
```

# Histograms

```python
a=np.array([27,43,5,21,93,67,82,14,79,10])
plt.hist(a)
plt.title("Histogram")
```

## Box plot using Matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data, vert=True, patch_artist=True);
```

# Pi Chart

```python
# data to plot
labels=("Python", "C++", "Ruby", "Java")
colours=("Green", "Yellow", "Red", "Blue")
sizes=[210,110,289,201]
explode=(0.2,0,0,0) # explode 1st slide

#plot
plt.pie(sizes, explode= explode, labels=labels, colors= colours, autopct='%1.1f%%', shadow=True )
plt.axis('equal')
plt.show()
```
